```python
'''
BLACKJACK Source code
'''

from random import shuffle                                   #Imports shuffle(), it can shuffle any ordered sequence
from IPython.display import clear_output                     #Imports clear_output()(only in IDE's having IPython)

#Assignments for organisation of a pack of one deck of cards.
#To play with more than one deck, make pack = pack * n, where n is the no. of decks.
suits = ['Hearts','Clubs','Spades','Diamonds']               
ranks = ['Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace']
value = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10,'Ace':1}
pack = []
for i in range(0,4):
    k = 0
    for j in range(i*13,i*13+13):
        pack.append((ranks[k],suits[i]))
        k += 1

                                         #----------------------------

def intro():
    '''
    To Introduce the game.
    INPUT:No input
    OUTPUT: Prints the introduction
    '''
    print('\n\n\n\n\n')
    print('\t --------   |                 --          --------  |    /   --------      --          --------  |    /  ')
    print('\t|        |  |               /    \       |          |   /        |       /    \       |          |   /   ')
    print('\t|        |  |              /      \      |          |__/         |      /      \      |          |__/    ')
    print('\t|--------   |             /--------\     |          |  \         |     /--------\     |          |  \    ')
    print('\t|        |  |            /          \    |          |   \        |    /          \    |          |   \   ')
    print('\t|        |  |           /            \   |          |    \       |   /            \   |          |    \  ')
    print('\t --------    --------  /              \   --------  |     \  ----   /              \   --------  |     \ ')
    print('\n\n\n\n\n')
    input('\t\t\tPress any key to continue')
    
    clear_output()                                           #Clears the output
    
    print(f"{'-'*50}BlackJack{'-'*50}")
    print('\nObjective:')
    print('"The object of blackjack is to get as close to 21 as possible, without going over." No! The object of')
    print('blackjack is to beat the dealer.  To beat the dealer the player must first not bust (go over 21) and second')
    print('either outscore the dealer or have the dealer bust.')
    print('\nRules:')
    print('    1.This version is limited to one player only.')
    print('    2.Blackjack may be played with one to eight decks of 52-card decks. This version is limited to one deck.')
    print('    3.Aces may be counted as 1 or 11 points,and tens and face cards count as ten points.')
    print('    4.The value of a hand is the sum of the point values of the individual cards.') 
    print('      The sum is 21 when:') 
    print('       --There is no ace, and all the cards sums up to 21.')
    print('       --There is an ace and some 10 value card in first turn. Then there is a BlackJack. This makes either')
    print('         dealer or player to win 2.5 times the bet.')
    print('    5.Player bets using chips. Each chip costs $1.')
    print('    6.After the players have bet, the dealer will give two cards to each player and two cards to himself. One')
    print('      of the dealer cards is dealt face up. The facedown card is called the "hole card".')
    print("    7.Only two choices are given to the player in this version. They are limited as compared to original")
    print('      version. These are- ')
    print('        Stand: Player stands pat with his cards. Once stand is called by the player then, the dealer')
    print("               hits until his cards' sum becomes 17 or greater than 17")
    print("        Hit: Player draws another card (and more if he wishes). If this card causes the player's")
    print('             total points to exceed 21 (known as "breaking" or "busting") then he loses.')
    print('    8.After the player has had his turn, the dealer will turn over his hole card.')
    print("    9.If the dealer goes over 21 points, then the player if didn't already bust will win.")
    print('    10.If the dealer does not bust, then the higher point total between the player and dealer will win.')
    print('    11.Except a BlackJack, you win or lose 2 times the bet money.')
    input('\n\t\t\tPress any key to play')
    
    clear_output()

                                         #----------------------------

class Game():
    '''
    To create the user objects to handle the game i.e. dealer,player.
    '''
                                                   #---------
    
    def __init__(self,name,tag,cards,chips = 0):
        '''
        Default Game class constructor.
        INPUT:Object attributes
        OUTPUT: No output
        '''
        self.name = name
        self.tag = tag
        self.cards = cards
        self.chips = chips
    
                                                    #---------

    def bet_stats(self,bet,chips,win = True):
        '''
        To manage the no. of chips.
        INPUT: Player's bet, Player's chips, and win indicates player's luck on the bet
        OUTPUT: Prints the status of chips.
        '''
        if win == True:                             #if player won the bet
            self.chips += bet
        else:                                       #if player lost the bet
            self.chips -= bet
        
        if type(self.chips) == float:               #if chips are of float type, convert it to integer
            self.chips = int(self.chips)
        
        print(f'\n\t\t\t\t\tNow, You have {self.chips} chip(s).')
         
        if self.chips == 0:                         #if chips have reduced to zero
            print("\n\t\t\t You have no chips now. You can't play another hand.")
            print('\n\t\t\t\t\t\tHave a nice day!')
        
        if self.chips < 0:                          #if player has been debted
            print("\n\t\t\t  Your chips are over.You are required to pay {self.chips} chip(s) within a week.")
            print("\t\t\tYour information is registered with our Casino's customer department.")
            print("\tWe will send an executive to your address by one week from today, to collect the debt if not payed till then.")
            print("\t      If you want more time,1.5% interest per day will be charged after the first week. Thank You!")
            print('\n\t\t\t\t\t\tHave a nice day!')

                                                    #---------
    
    def printer(self,cards):
        '''
        Displays the user's cards side by side depending on the moves and conditions.
        INPUT:User's cards
        OUTPUT: Prints the cards
        '''
        num_list = []                               #list to store ranks of the cards       
        suit_list = []                              #list to store suits of the cards
        
        for i in range(len(self.cards)):            #Loop to fill the above lists
          #if rank is either 'Jack','Queen','King' or 'Ace, append its initial letter to the num_list
            if cards[i][0] in list(value.keys())[9:]:      
                num_list.append(self.cards[i][0][0])
          #else append the ranks as they are  
            else:
                num_list.append(value[self.cards[i][0]])
            
          #append the initials of suits to suit_list  
            suit_list.append(self.cards[i][1][:1])     
          
          #An object is given a tag which determines whether the card is required to be shown or not. It has
          #significance in the dealer's case as he/she has to hide the second card in some situations.
          #Default tag provided to the dealer is 'd'. Whenever the tag is changed to 'D', it gives the green 
          #signal to the dealer's second card to be shown.
            
          #if dealer's second card is not needed to be shown   
            if self.tag == 'd':   
                num_list.append('\\')
                suit_list.append('o')
                break
        
     #The following statements in this method, prints the required cards calling dynamic_cards()    
        print('\n\t\t',end='')
        print('-----------    '*len(self.cards))
        print('\t\t',end='')
        print(dynamic_cards(num_list,0,8))
        for i in range(0,3):
            print('\t\t',end='')
            if i == 1:
                print(dynamic_cards(suit_list,4,4))
            else:
                print("|         |    "*len(self.cards))
        print('\t\t',end='')
        print(dynamic_cards(num_list,8))
        print('\t\t',end='')
        print('-----------    '*len(self.cards))
    
                                                    #---------
    
    def card_sum(self,cards):
        '''
        Calculates the sum of the user's cards ranks.
        INPUT:User's cards
        OUTPUT: Returns the sum
        '''
        sum_var=0                                    #Stores the sum
        
        for i in range(len(self.cards)):             #Loop to calculate the sum              
            sum_var += value[self.cards[i][0]]
        
        for i in range(len(self.cards)):
          #if any card is an Ace, adjust the sum using its value as either 1 or 11 to make the sum maximum and <= 21
            if self.cards[i][0] == 'Ace' and sum_var + 10 <= 21:
                sum_var += 10
        
        return sum_var                               #Returns the sum
    
                                                    #---------
    
    def blackJack(self):
        '''
        Checks if the user has a blackJack or not.
        INPUT: No input
        OUTPUT: Returns a boolean
        '''
        for i in range(0,2):                         #Loop to check for blackJack  
          #if one of the cards is Ace, and the other card has a value 10, then return True  
            if self.cards[i][0] == 'Ace':
                if i == 0:
                    if value[self.cards[1][0]] == 10:
                        return True
                if i == 1:
                    if value[self.cards[0][0]] == 10:
                        return True
        
        return False                                 #if not a blackJack, return False           


                                         #----------------------------

def dynamic_cards(value_list,space_val1=0,space_val2=0):
    '''
    Creates a string for user's cards parts(which contain value or suit), so as to achieve side by side
    printing of cards.
    INPUT: Either suit_list or num_list, 2 variables assigned to 0 by default indicates where to print whitespaces
           depending on values or suits
    OUTPUT: Returns the created string
    '''
    to_join_list = []                                #List to join cards in sequence and create one string  
    print_it = ''                                    #The string to store the sequence
    
    for i in range(len(value_list)):                 #Loop to create print_it
      #if rank is 10, reduce one space so that the string is adjusted properly for printing  
        if value_list[i]==10:
            if space_val1 == 0: to_join_list.append(f"|{' '*space_val1}{value_list[i]}{' '*(space_val2 - 1)}|    ")
            else: to_join_list.append(f"|{' '*(space_val1 - 1)}{value_list[i]}{' '*space_val2}|    ")      
        else:
            to_join_list.append(f"|{' '*space_val1}{value_list[i]}{' '*space_val2}|    ")
        
        print_it = ''.join(to_join_list)             #print_it is created  
    
    return print_it                                  #Return the string

                                         #----------------------------

def table(dealer,player,sum_list = None,move = None,first = True):
    '''
    Displays the table made up by cards,dealer,player and their cards' sum.
    INPUT: Dealer and Player object, list of their cards' sum, move made by player, and a variable which indicates
           whether first hand is being played or not
    OUTPUT: Prints the table
    '''
    print(f"\n\t{'-'*40}Dealer{'-'*40}")           
    dealer.printer(dealer.cards)                     #Calls Game.printer() to print dealer's cards      
    
  #if move is either hit or stand, or first hand is being played
    if move == 'h' or move == 's' or first == True:
      #if dealer is required to show his/her cards  
        if dealer.tag == 'D':
            print("\n\t\t\t\t\tDealer's card Sum: ",sum_list[0])
        print('\n\n\n')
        print("\n\t\t\t\t\tYour's card Sum: ",sum_list[1])
    else: print('\n\n\n')
    
    player.printer(player.cards)                     #Calls Game.printer() to print player's cards
    print(f"\n\t{'-'*40}{player.name}{'-'*40}")

                                         #----------------------------

def hit_stand(dealer,player):
    '''
    Recieves the dealer's cards sum and player's cards sum separately by calling card_sum()and
    then returns them combined in a list.
    INPUT: Dealer and Player object
    OUTPUT: returns the created string
    '''
    d_sum = dealer.card_sum(dealer.cards)            #Dealer's card sum    
    p_sum = player.card_sum(player.cards)            #Player's card sum
    
    return [d_sum,p_sum]                             #Returns sum as a list

                
                                         #----------------------------

def busted(dealer,player,sum_list,bet):
    '''
    Check whether dealer or player is busted or not, updates the chips by calling Game.bet_stats()
    and prints the result.
    INPUT: Dealer and Player object, list of their cards' sum, player's bet
    OUTPUT: Prints the result
    '''
  #if dealer's cards' sum is greater than 21
    if sum_list[0] > 21:    
        print('\n\t\t\t\t\tDealer is busted !!!')
        print('\n\t\t\t\t\tYou won this hand')
        player.bet_stats(bet,player.chips)
  
 #if player's cards' sum is greater than 21  
    elif sum_list[1] > 21:
        print('\n\t\t\t\t\tYou are busted !!!')
        print('\n\t\t\t\t\tDealer won this hand')
        player.bet_stats(bet,player.chips,False)

                                         #----------------------------

def who_won(dealer,player,sum_list,bet,bust):
    '''
    Check whether dealer or player won, updates the chips by calling Game.bet_stats() and prints the
    result
    INPUT: Dealer and Player object, list of their cards' sum, player's bet
    OUTPUT: Prints the result
    '''
  #if dealer's cards' sum is greater than player's cards' sum and no one is busted
    if sum_list[0] > sum_list[1] and bust == False:
        print('\n\t\t\t\t\tDealer won this Hand !!!')
        print('\n\t\t\t\t\tYou lost the bet.')
        player.bet_stats(bet,player.chips,False)

  #if dealer's cards' sum is equal to player's cards' sum and no one is busted  
    elif sum_list[0] == sum_list[1] and bust == False:
        print('\n\n\t\t\t\t\t No one won this hand.')
        print('\n\t\t\t\t\tYou lose half the chip(s).')
        bet = bet/2
        player.bet_stats(bet,player.chips,False)

  #if player's cards' sum is greater than dealer's cards' sum and no one is busted
    elif sum_list[0] < sum_list[1] and bust == False:    
        print('\n\n\t\t\t\t\tYou won this Hand !!!')
        player.bet_stats(bet,player.chips) 
            
                                         #----------------------------

def play_game():
    '''
    This is the main function which creates the game environment and controls the use of all the other
    functions and the Game() class.
    INPUT: No input
    OUTPUT: Output is not specific
    '''
    intro()                                          #Calls intro() to print introduction                    
    
    shuffle(pack)                                    #Shuffles the pack
    
    name = input("What's your name?: ").capitalize() #Inputs name
    
    while True:                                      #Loop to ensure that the player enters valid input for chips
        try:
            chips = int(input(f'{name}, How many chip(s) you have? '))
        except:
            print('\n Enter something Valid.')
            continue
        else:
            break
    
    clear_output()
    
  #Assignments to be used in the following loop   
    i = 0                                            #Stores a cards index in the pack 
    move = None                                      #Stores the player's move
    hand = 'Y'                                       #Variable to check if player wants to replay or not
    
  #Loop to replay the game if user wants and all cards are not showed up  
    while hand != 'N' and i <= 48:                              
      #Assignments which are required to be resetted on each hand      
        bet = 0                                      #Stores the bet on each hand 
        first = True                                 #Boolean to indicate that first hand is being played or not
      #Changes to 'd' if its dealer's turn to hit, doesn't ask for move  
        dealer_hit = None                            
        over = False                                 #Boolean to indicate that a hand is over or not
        bust = False                                 #Boolean to indicate that a user is busted or not
        bj = False                                   #Boolean to indicate that a user has a blackJack or not
      #Changes to 'N', when few cards are left and player doesn't wish to continue
        last = 'Y'                                   
        
      #if cards in the pack are finishing up and only few are left  
        if i > 44:
            print("\n Few cards left in the pack. You may or may not hit at any particular move.")
            last = input(" Do you want to continue?('Y' or 'N'): ").upper()
            clear_output()
            if last == 'N': break
      
      #if cards in the pack are not enough to play another hand 
        if i > 48:
            print("Can't play another hand. Not enough cards. See you next time!")
            print('\n\t\t\t\t\tPress any key')
            break
      
        print(f'You have {chips} chip(s).')
        
        while True:                                  #Loop to ensure that the player enters valid input for bet                        
            
            bet = int(input('Enter the chip(s) you want to bet: '))
            
            if type(bet) != int:                     #if bet is not integer type, continue
                print("\nEnter something valid.")
                continue
            if bet > chips:                          #if bet is more than available chips, continue
                print("\nCan't bet more than available chip(s). Re- Enter")
                continue
            else: break
      
      #Object player created with two cards initially  
        player = Game(name,'p',[pack[i],pack[i+1]],chips)
      #Object dealer created with two cards initially
        dealer = Game('Dealer','d',[pack[i+2],pack[i+3]])                     
        
        clear_output()
        
        print(f'\n\t\t\t\t\t  {bet} chip(s) are on bet.')
        
      #if player has a blackJack, then dealer cards are required to be shown
        if player.blackJack == True: dealer.tag = 'D'  
        
      #List of cards' sums recieved in the sum_list returned by hit_stand
        sum_list = hit_stand(dealer,player)
        
      #Display the table with initial cards  
        table(dealer,player,sum_list)
        
      #if player has a blackJack, print the result, update the chips by calling Game.bet_stats(), and take to the end  
        if player.blackJack() == True:
            print('\n\t\t\t\t\t  You have a BlackJack.')
            print('\t\t\t\tCONGRATULATIONS! You have won 2.5 times the bet.')
            bet = bet + bet/2
            player.bet_stats(bet,player.chips)
            bust = True
            over = True
            bj = True
        
        i = i+4                                      #index is increased by 4, as 4 cards have been drawn already               
          
        while over == False:                         #Loop to continue if the hand is not over yet 
            if dealer_hit != 'd':
          #Loop to ensure that player enters a valid value for move  
                  while True:
                        move = input("\nCall your move('H' for HIT OR 'S' for STAND): ")
                        if move.lower() not in ['h','s']:
                            print('\n Not a Valid move. Re-enter')
                            continue
                        else:
                            break
            
            if move.lower() == 'h':                  #if player hits     
              #if cards are over and player has still hit  
                if i > 50 and last != 'N':
                    print("\n\t\t\t\t You can't hit more.")
                    continue
                else:
              #else append the next card to the player's cards and update the sum and card index
                    player.cards.append(pack[i])
                    sum_list = hit_stand(dealer,player)
                    i += 1

            else:                                    #if player stands         
                dealer.tag = 'D'                     #Dealer tag changed so as to show second card along
                over = True                          #Once stand is called, no next move for player
                
                sum_list = hit_stand(dealer,player)  #update the sum
                
              #if dealer's cards' sum is less than 17, then he/she hits until the sum become >=17  
                if sum_list[0] < 17:
                    if i > 50: pass                  #if cards in the pack are over, can't hit
                    else:
                        dealer.cards.append(pack[i]) #Append the next card to dealer's cards
                        sum_list = hit_stand(dealer,player)
                        dealer_hit = 'd'             #Don't ask for move as the loop is to be repeated
                        over = False                 #Repeat the loop
                        i += 1                       #Update the card index
                
                else: over = True                    #else dealer can't hit as his/her cards' sum > 17
            
            clear_output()
          
          #if anyone is busted, then dealer has to show his second card and go to the end
            if sum_list[0] > 21 or sum_list[1] > 21:                              
                dealer.tag = 'D'
                bust = True
                over = True
            
          #if dealer has a BlackJack, then show the second card  
            if dealer.blackJack() == True: dealer.tag = 'D'
            
            table(dealer,player,sum_list,move,False) #Display the table after a move      
            
          #if dealer has a BlackJack, print the result, update the chips, and go to the end  
            if dealer.blackJack() == True:
                print('\n\t\t\t\tDealer has a BlackJack. You lost 1.5 times the bet.')
                bet = bet + bet/2
                player.bet_stats(bet,player.chips,False)
                bust = True
                over = True
                bj = True
          
            busted(dealer,player,sum_list,bet)
                   
        else:                                        #if dealer and player has completed their moves              
            
            if not bj:                               #if no one has a blackJack
                who_won(dealer,player,sum_list,bet,bust) 
            
        chips = player.chips                         #Update the chips variable beig used in this function
        
        if i >= 51:                                  #If all cards are shown, end the game
            print('\n\t\t\t\t    All cards showed up. Game Over !!!')
            print('\n\t\t\t\t\t\tHave a nice day!')
            input('\n\t\t\t\tPress any key')
            break
        
        if chips <= 0: break                          #If chips have reduced to 0 or player in debt, end the game
        
        else:                                         #else, ask for playing another hand
          #Loop to ensure that player enters a valid input for hand  
            while True:
                hand = input("\n\t\t Wanna play another Hand?('Y' or 'N'): ").upper()                
                if  hand not in ['Y','N']:
                    print('\n\t\t   Not a valid answer. Re-enter')
                    continue
                else:
                    break
        
        clear_output()

#--------------------------------------------------------------------------------------------------------------------------------------
```

```python
play_game()                                              #call statement to start the game
```
